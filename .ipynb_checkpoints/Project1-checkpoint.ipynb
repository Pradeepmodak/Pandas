```python
def extract_episode(txt):
    check=False
    data="" 
    for i in txt:
        if(i=="("):
           check=True
           continue 
        if(i==")" and check==True):
           check=False 
           return data
           break
        if(check):
            data+=i
```

```python
eps=extract_episode(df.loc[2]['Title'])
eps
```

```python
df['Episodes']=df["Title"].apply(extract_episode)
df
```

```python
df["Episodes"]=df["Episodes"].str.replace(" eps","")
df
```

```python
def extract_timestamp(txt):
    j=-1
    for i in range(0,len(txt),1):
        if txt[i]==")":
            j=i+1
            break
    data=""
    for i in range(j,j+19,1):
        data+=txt[i] 
    return data   
```

```python
df["TimeStamp"]=df["Title"].apply(extract_timestamp)
df
```

```python
from datetime import datetime

def calculate_months(duration_str):
    # Split the string on the dash
    parts = duration_str.split("-")
    if len(parts) != 2:
        return 0   # invalid format

    start_str = parts[0].strip()   # "Apr 2009"
    end_str = parts[1].strip()     # "Jul 2010"

    # Define the format to parse month and year
    fmt = "%b %Y"   # %b = abbreviated month name, %Y = 4-digit year

    # Parse into datetime objects
    start_date = datetime.strptime(start_str, fmt)
    end_date = datetime.strptime(end_str, fmt)

    # Calculate total months
    total_months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month) + 1
    return total_months
```

```python
df["Months"]=df["TimeStamp"].apply(calculate_months)
```
