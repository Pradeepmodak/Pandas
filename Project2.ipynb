```python
df[df['political_leader'].isna()]['country']
```

```python
# How many countries are there having Republic in their full name
```

```python
count=0
def counting(txt):
    global count 
    if 'republic' in txt.lower():
        count+=1;
    return txt
df['country_long'].apply(counting)
print(count)
```
